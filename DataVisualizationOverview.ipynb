```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,10)
```

```python
x
```

```python
y = x**2
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, '*')
```

```python
plt.plot(x,y, 'r--')
plt.xlim(0,4)
plt.ylim(0,10)
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label')
```
